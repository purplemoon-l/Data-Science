```python
def bias_desc_cleaning(df, col, threshold = 500):
    l = df[col].value_counts()
    less_than_threshold = []
    for key in l.keys():
        if l[key] < threshold:
            less_than_threshold.append(key)

    bins = [ 0,1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000]
    plt.hist(l, bins, color ='purple', alpha = 0.7)

    plt.show()

    bins = [ 9000,19000, 29000, 39000, 49000, 59000, 69000, 79000]
    plt.hist(l, bins, color ='purple', alpha = 0.7)
    
    plt.show()
    
    print("Before Cleaning Length of Values", len(df))
    print("Unique Values", len(df[col].value_counts()),"\n\n\n")
    
    df_cleaned = df[~df[col].isin(less_than_threshold)]
    print("After Cleaning Length of Values", len(df_cleaned))
    print("Unique Values", len(df_cleaned[col].value_counts()),"\n\n\n")
    return df_cleaned

    
df_cleaned = bias_desc_cleaning(df, "BIAS_DESC")



merging_bias_description = {"Anti-Gay (Male)":"Anti-Queer",
                            "Anti-Lesbian (Female)":"Anti-Queer",
                            "Anti-Transgender":"Anti-Queer",
                            "Anti-Lesbian, Gay, Bisexual, or Transgender (Mixed Group)" : "Anti-Queer",
                            "Anti-Bisexual" : "Anti-Queer",
                            "Anti-Other Race/Ethnicity/Ancestry":"Anti-Other Religion",
                            "Anti-Multiple Religions, Group": "Anti-Other Religion",
                            "Anti-Multiple Races, Group":"Anti-Other Religion",
                            "Anti-Catholic": "Anti-Christianity",
                            "Anti-Protestant": "Anti-Christianity"
                           }

def merge_bias_description(df,col, merging_dict):
    print("Before", df[col].value_counts())
    df2=df.replace({col: merging_dict})
    print("\n\nAfter", df2[col].value_counts())
    
merge_bias_description(df_cleaned, "BIAS_DESC", merging_bias_description)
```

```python
df_cleaned = bias_desc_cleaning(df_cleaned, "OFFENSE_NAME")
```

```python
import matplotlib.pyplot as plt
df= df_cleaned
for i in list(df["STATE_NAME"].unique()):
    df1 = df[df["STATE_NAME"]==i]
    if len(df1)>500 :
        l = []
        for j in list(df["DATA_YEAR"].unique()):
            l.append(len(df1[df1["DATA_YEAR"]==j]))
        plt.plot(df["DATA_YEAR"].unique(), l, label = i)
        plt.legend()
        plt.show()
```
